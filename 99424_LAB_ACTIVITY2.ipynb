```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn #ติดตั้ง sklearn ด้วยคาสั่ง pip install scikit-learn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset =pd.read_csv('/content/drive/MyDrive/99424_1/GradingSystem.csv')
dataset.head(10)
```

```python
x = dataset['Math']
y = dataset['CGPA']
```

```python
plt.scatter(x,y,color='red')
plt.xlabel('MATH', fontsize=14)
plt.ylabel('CGPA', fontsize=14)
plt.show()
```

```python
x = dataset.iloc[:, 1].values.reshape(-1, 1) #อาร์เรย์ตัวแปรอิสระ
y = dataset.iloc[:, 2].values.reshape(-1, 1) #อาร์เรย์ตัวแปรตาม (row,column)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y,
                 train_size=0.7, test_size=0.3,random_state=0)
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train,y_train)
```

```python
y_predicted = regression_model.predict(x_test)
y_predicted
```

```python
y_test
```

```python
rmse = mean_squared_error(y_test, y_predicted)
r2_score = r2_score(y_test,y_predicted)
```

```python
print('The intercept is:', regression_model.intercept_)
print('The coefficient is:' ,regression_model.coef_)
print('The rmse is:',rmse)
print('The r2_score is:',r2_score)
```
