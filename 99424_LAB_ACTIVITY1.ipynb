```python
#1.1.1.การนาเข้าข้อมูล
import numpy as np
import pandas as pd
#สร้างข้อมูลนักศึกษา 5 คน และคะแนนสอบ 5 รายการ
students = np.array([
     ["S001", "Anan", 80, 75, 85, 90, 88],
     ["S002", "Boon", 65, 70, 60, 68, 72],
      ["S003", "Chai", 90, 88, 92, 95, 94],
      ["S004", "Dara", 55, 60, 58, 62, 59],
      ["S005", "Wit", 78, 82, 80, 85, 83] ])
```

```python
#1.1.2.การแปลง array ให้เป็น data frame
#การแปลง Array เป็น DataFrame
mydata = pd.DataFrame(students, columns=[
     "StudentID", "Name", "Quiz", "Midterm", "Assignment",
     "Project", "Final"
     ])
mydata
```

```python
midterm_mean = data_excel["Midterm"].mean()
missing_rows = data_excel[data_excel["Midterm"].isnull()]
print("แถวที่มีค่า Midterm หาย")
print(missing_rows)
print("\nควรเติมค่า =", round(midterm_mean, 2))
```

```python
import numpy as np
numeric_cols = data_excel.select_dtypes(include=np.number)
outliers = data_excel[(numeric_cols > 100).any(axis=1)]
print(outliers)
```

```python
import pandas as pd
data_excel1 = pd.read_excel("/content/sample_data/students_1.xlsx")
data_excel1.head()
```

```python
data_excel1.describe()
```

```python
import matplotlib.pyplot as plt
data_excel1.set_index("Name")[["Quiz", "Midterm", "Assignment", "Project", "Final"]].plot(
     kind="bar",
     figsize=(12,6)
     )
plt.title("Student Scores")
plt.xlabel("Student Name")
plt.ylabel("Scores")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.pie( data_excel1["Final"],
        labels=data_excel1["Name"],
         autopct='%1.1f%%'
)
plt.title("Final Score Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
data_excel1.boxplot(column=["Quiz", "Midterm",
                            "Assignment", "Project", "Final"])
plt.title("Boxplot of Student Scores")
plt.ylabel("Scores")
plt.show()
```
